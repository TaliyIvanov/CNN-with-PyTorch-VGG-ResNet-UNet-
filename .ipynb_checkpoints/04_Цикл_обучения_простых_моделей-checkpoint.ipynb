````python
def read(dataset):
    if dataset == 'training':
        path_img = '/content/sample_data/MNIST/raw/train-images-idx3-ubyte'
        path_lbl = '/content/sample_data/MNIST/raw/train-labels-idx1-ubyte'

    elif dataset == 'testing':
        path_img = '/content/sample_data/MNIST/raw/t10k-images-idx3-ubyte'
        path_lbl = '/content/sample_data/MNIST/raw/t10k-labels-idx1-ubyte'

    else:
        raise ValueError("dataset must be 'testing' or 'training'")

    with open(path_lbl, 'rb') as f_lable:
        _, size = struct.unpack('>II', f_lable.read(8))
        lbl = array('b', f_lable.read())

    with open(path_img, 'rb') as f_img:
        _, size, rows, cols = struct.unpack('>IIII', f_img.read(16))
        img = array('B', f_img.read())

    return lbl, img, size, rows, cols
````

````python
def write_dataset(labels, data, size, rows, cols, output_dir):

    classes = {i: f'class_{i}' for i in range(10)}

    output_dirs = [
        path.join(output_dir, classes[i])
        for i in range(10)
    ]
    for dir in output_dirs:
        if not path.exists(dir):
            os.makedirs(dir)

    # write data
    for (i, label) in enumerate(labels):
        output_filename = path.join(output_dirs[label], str(i) + '.jpg')
        print('writing ' + output_filename)

        with open(output_filename, 'wb') as h:
            data_i = [
                data[(i*rows*cols + j*cols) : (i*rows*cols + (j+1)*cols)]
                for j in range(rows)
            ]
            data_array = np.asarray(data_i)

            im = Image.fromarray(data_array)
            im.save(output_filename)
````

````python
# определяем на каком устройстве будем проводить обучение
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
````

#### Создание Dataset

````python
# используем ранее созданные пользовательский класс
class MNISTDataset(Dataset):
    def __init__(self, path, transform=None):
        self.path = path
        self.transform = transform

        self.len_dataset = 0 # длина датасет
        self.data_list = [] # список кортежей путей до файла и позиции в onehot векторе

        # итерируемся по папке с основными файлами
        for path_dir, dir_list, file_list in os.walk(path):
            if path_dir == path:
                self.classes = sorted(dir_list)
                self.class_to_idx = {
                    cls_name: i for i, cls_name in enumerate(self.classes)
                    }
                continue

            cls = path_dir.split('/')[-1]

            for name_file in file_list:
                file_path = os.path.join(path_dir, name_file)
                self.data_list.append((file_path, self.class_to_idx[cls]))

            self.len_dataset += len(file_list)

    def __len__(self):
        return self.len_dataset

    def __getitem__(self, index):
        file_path, target = self.data_list[index]
        sample = np.array(Image.open(file_path))

        if self.transform is not None:
            sample = self.transform(sample)

        return sample, target
````

````python
# Преобразования для изображений
transform = v2.Compose(
    [
        v2.ToImage(),
        v2.ToDtype(torch.float32, scale=True),
        v2.Normalize(mean=(0.5, ), std=(0.5, ))
    ]
)
````

````python
# создание датасетов
train_data = MNISTDataset('/content/mnist/training', transform=transform)
test_data = MNISTDataset('/content/mnist/testing', transform=transform)
````

````python
# разделение трейновой части на трейн и валидацию
train_data, val_data = random_split(train_data, [0.7, 0.3])
````

````python
# создание загрузчиков
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
val_loader = DataLoader(val_data, batch_size=64, shuffle=False)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)
````

#### Создание модели

````python
class MyModel(nn.Module):
    def __init__(self, input, output):
        super().__init__()
        self.layer_1 = nn.Linear(input, 128)
        self.layer_2 = nn.Linear(128, output)
        self.act = nn.ReLU()

    def forward(self, x):
        x = self.layer_1(x)
        x = self.act(x)
        out = self.layer_2(x)
        return out
````

````python
# создаем и отправляем на cuda
model = MyModel(784, 10).to(device)
````

````python
# Проверяем правильность построения модели
input = torch.rand([16, 784], dtype=torch.float32).to(device)

out = model(input)
out.shape    # (16,10)
````

````python
# выбираем функцию потерь и оптимизатор градиентного спуска
loss_model = nn.CrossEntropyLoss()
opt = torch.optim.Adam(model.parameters(), lr=0.001)
````

#### Тренировка модели

````python
# Цикл обучения

#     1. Тренировка модели
#        - Данные
#        - Прямой проход + расчет ошибки модели
#        - Обратный проход
#        - Шаг оптимизации

#     2. Расчет значения метрики
#     3. Сохранение значения функции потерь и метрики

#     4. Проверка модели (валидация)
#        - Данные
#        - Прямой проход + расчет ошибки модели
#     5. Расчет значения метрики
#     6. Сохранения значения функции потерь и метрики
````

**Цикл обучения (итерация по эпохам):**

```
for epoch in range(EPOCHS):
```

**Внутри каждой эпохи происходят следующие шаги:**

1.  **Тренировка модели (`model.train()`):**

    *   Модель переводится в режим обучения, позволяющий обновлять веса.
    *   Итерация по тренировочным данным, разделенным на батчи (`train_loader`).
    *   **Для каждого батча:**
        *   Данные (`x`) преобразуются к подходящему для модели формату и перемещаются на устройство (GPU/CPU).
        *   Метки (`targets`) также преобразуются к подходящему формату (one-hot encoding) и перемещаются на устройство.
        *   Выполняется прямой проход через модель, получая предсказания (`pred`).
        *   Вычисляется функция потерь (`loss`) между предсказаниями и фактическими метками.
        *   Выполняется обратный проход (`loss.backward()`) для расчета градиентов.
        *   Обновляются веса модели (`opt.step()`).
        *   Рассчитывается среднее значение функции потерь (`mean_train_loss`) и точность (`running_train_acc`) для текущей эпохи обучения.
        *   Прогресс обучения отображается через `tqdm` (прогресс бар).
2.  **Сохранение результатов тренировки:**
    *   Средняя функция потерь (`mean_train_loss`) и точность (`running_train_acc`) для текущей эпохи добавляются в соответствующие списки (`train_loss`, `train_acc`).

3.  **Валидация модели (`model.eval()`):**

    *   Модель переводится в режим оценки, отключающий градиенты и позволяющий оценивать производительность модели.
    *   Итерация по валидационным данным, разделенным на батчи (`val_loader`).
    *   **Для каждого батча:**
        *   Данные (`x`) и метки (`targets`) обрабатываются аналогично этапу тренировки.
        *   Выполняется прямой проход через модель, получая предсказания (`pred`).
        *   Вычисляется функция потерь (`loss`).
        *   Рассчитывается средняя функция потерь (`mean_val_loss`) и точность (`running_val_acc`) для текущей эпохи валидации.
4.  **Сохранение результатов валидации:**
    *   Средняя функция потерь (`mean_val_loss`) и точность (`running_val_acc`) для текущей эпохи добавляются в соответствующие списки (`val_loss`, `val_acc`).

5.  **Вывод результатов:**
    *   Результаты текущей эпохи (функция потерь и точность на тренировочных и валидационных данных) выводятся в консоль.

**Ключевые моменты:**

*   **Батчирование:** Данные обрабатываются небольшими группами (батчами) для эффективной работы с памятью и вычислениями.
*   **Прямой и обратный проходы:** Прямой проход вычисляет предсказания, обратный проход вычисляет градиенты для обновления весов.
*   **Оптимизатор (`opt`):** Алгоритм, используемый для обновления весов модели на основе градиентов.
*   **Функция потерь (`loss_model`):** Функция, измеряющая разницу между предсказаниями и фактическими метками.
*   **Точность (accuracy):** Метрика, показывающая, насколько хорошо модель классифицирует данные.
*   **Тренировочные и валидационные данные:** Тренировочные данные используются для обучения модели, валидационные данные используются для оценки производительности модели на "невидимых" данных.

Этот алгоритм представляет собой стандартный процесс обучения модели, и он может быть адаптирован для решения различных задач машинного обучения, изменяя архитектуру модели, функцию потерь и оптимизатор.


````python
EPOCHS = 50
train_loss = []
train_acc = []
val_loss = []
val_acc = []

# Цикл обучения
for epoch in range(EPOCHS):

    # 1. Тренировка модели
    model.train()
    running_train_loss = []
    true_answer = 0
    # добавим трейн луп, чтобы видеть прогресс обучения модели
    train_loop = tqdm(train_loader, leave=False)
    for x, targets in train_loop:
        # - Данные
        # (batch.size, 1, 28, 28) --> (batch.size, 784)
        x = x.reshape(-1, 28*28).to(device)
        # (batch.size, int) --> (batch.size, 10), dtype.float32
        targets = targets.reshape(-1).to(torch.int32)
        targets = torch.eye(10)[targets].to(device)

        # - Прямой проход + расчет ошибки модели
        pred = model(x)
        loss = loss_model(pred, targets)

        # - Обратный проход
        opt.zero_grad()
        loss.backward()
        # - Шаг оптимизации
        opt.step()

        running_train_loss.append(loss.item())
        mean_train_loss = sum(running_train_loss)/len(running_train_loss)

        true_answer += (pred.argmax(dim=1) == targets.argmax(dim=1)).sum().item()

        train_loop.set_description(f"Epoch [{epoch+1}/{EPOCHS}], train_loss={mean_train_loss:.4f}")


    # 2. Расчет значения метрики
    running_train_acc = true_answer / len(train_data)

    # 3. Сохранение значения функции потерь и метрики
    train_loss.append(mean_train_loss)
    train_acc.append(running_train_acc)

    # 4. Проверка модели (Валидация)
    model.eval()
    with torch.no_grad():
        running_val_loss = []
        true_answer = 0
        for x, targets in val_loader:
            # - Данные
            # (batch.size, 1, 28, 28) --> (batch.size, 784)
            x = x.reshape(-1, 28*28).to(device)
            # (batch.size, int) --> (batch.size, 10), dtype.float32
            targets = targets.reshape(-1).to(torch.int32)
            targets = torch.eye(10)[targets].to(device)

            # - Прямой проход + расчет ошибки модели
            pred = model(x)
            loss = loss_model(pred, targets)

            running_val_loss.append(loss.item())
            mean_val_loss = sum(running_val_loss)/len(running_val_loss)

            true_answer += (pred.argmax(dim=1) == targets.argmax(dim=1)).sum().item()

        # 2. Расчет значения метрики
        running_val_acc = true_answer / len(val_data)

        # 3. Сохранение значения функции потерь и метрики
        val_loss.append(mean_val_loss)
        val_acc.append(running_val_acc)

        print(f"Epoch [{epoch+1}/{EPOCHS}], train_loss={mean_train_loss:.4f}, train_acc={running_train_acc:.4f}, val_loss={mean_val_loss:.4f}, val_acc={running_val_acc:.4f}")
````

````python
# График отображения функции потерь на тренировке и валидации
plt.plot(train_loss)
plt.plot(val_loss)
plt.legend(['loss_train', 'loss_val'])
plt.show
````

#### Подготовка данных

````python
class DatasetReg(Dataset):
    def __init__(self, path, transform=None):
        self.path = path
        self.transform = transform

        self.list_name_file = os.listdir(path)
        if 'coords.json' in self.list_name_file:
            self.list_name_file.remove('coords.json')

        self.len_dataset = len(self.list_name_file)

        with open(os.path.join(self.path, 'coords.json'), 'r') as f:
            self.dict_coords = json.load(f)

    def __len__(self):
        return self.len_dataset

    def __getitem__(self, index):
        name_file = self.list_name_file[index]
        path_img = os.path.join(self.path, name_file)

        img = Image.open(path_img)
        coord = self.dict_coords[name_file]

        if self.transform:
            img = self.transform(img)
            coord = torch.tensor(coord, dtype=torch.float32)


        return img, coord
````

````python
transform = v2.Compose(
    [
        v2.ToImage(),
        v2.ToDtype(torch.float32, scale=True),
        v2.Normalize(mean=(0.5, ), std=(0.5, ))
    ]
)
````

#### Создание модели для задачи регресии

````python
class MyModel(nn.Module):
    def __init__(self, input, output):
        super().__init__()
        self.layer_1 = nn.Linear(input, 128)
        self.layer_2 = nn.Linear(128, output)
        self.act = nn.ReLU()

    def forward(self, x):
        x = self.layer_1(x)
        x = self.act(x)
        out = self.layer_2(x)
        return out
````

````python
# создаем и отправляем на cuda
model = MyModel(64*64, 2).to(device)
````

````python
# Проверяем правильность построения модели
input = torch.rand([16, 64*64], dtype=torch.float32).to(device)

out = model(input)
out.shape    # (16,2)
````

````python
# выбираем функцию потерь и оптимизатор градиентного спуска
loss_model = nn.MSELoss()
opt = torch.optim.Adam(model.parameters(), lr=0.001)
````

#### Цикл обучения

````python
train_loss = []
train_acc = []
val_loss = []
val_acc = []
````

````python
EPOCHS = 20

# Цикл обучения
for epoch in range(EPOCHS):

    # 1. Тренировка модели
    model.train()
    running_train_loss = []
    true_answer = 0
    # добавим трейн луп, чтобы видеть прогресс обучения модели
    train_loop = tqdm(train_loader, leave=False)
    for x, targets in train_loop:
        # - Данные
        # (batch.size, 1, 64, 64) --> (batch.size, 64*64)
        x = x.reshape(-1, 64*64).to(device)
        targets = targets.to(device)

        # - Прямой проход + расчет ошибки модели
        pred = model(x)
        loss = loss_model(pred, targets)

        # - Обратный проход
        opt.zero_grad()
        loss.backward()
        # - Шаг оптимизации
        opt.step()

        running_train_loss.append(loss.item())
        mean_train_loss = sum(running_train_loss)/len(running_train_loss)

        true_answer += (torch.round(pred) == targets).all(dim=1).sum().item()

        train_loop.set_description(f"Epoch [{epoch+1}/{EPOCHS}], train_loss={mean_train_loss:.4f}")


    # 2. Расчет значения метрики
    running_train_acc = true_answer / len(train_set)

    # 3. Сохранение значения функции потерь и метрики
    train_loss.append(mean_train_loss)
    train_acc.append(running_train_acc)

    # 4. Проверка модели (Валидация)
    model.eval()
    with torch.no_grad():
        running_val_loss = []
        true_answer = 0
        for x, targets in val_loader:
            # - Данные
            # (batch.size, 1, 64, 64) --> (batch.size, 64*64)
            x = x.reshape(-1, 64*64).to(device)
            targets = targets.to(device)

            # - Прямой проход + расчет ошибки модели
            pred = model(x)
            loss = loss_model(pred, targets)

            running_val_loss.append(loss.item())
            mean_val_loss = sum(running_val_loss)/len(running_val_loss)

            true_answer += (torch.round(pred) == targets).all(dim=1).sum().item()

        # 2. Расчет значения метрики
        running_val_acc = true_answer / len(val_set)

        # 3. Сохранение значения функции потерь и метрики
        val_loss.append(mean_val_loss)
        val_acc.append(running_val_acc)

        print(f"Epoch [{epoch+1}/{EPOCHS}], train_loss={mean_train_loss:.4f}, train_acc={running_train_acc:.4f}, val_loss={mean_val_loss:.4f}, val_acc={running_val_acc:.4f}")
````

### Мои опыты

````python
class DatasetReg(Dataset):
    def __init__(self, path, transform=None):
        self.path = path
        self.transform = transform

        self.list_name_file = os.listdir(path)
        if 'coords.json' in self.list_name_file:
            self.list_name_file.remove('coords.json')

        self.len_dataset = len(self.list_name_file)

        with open(os.path.join(self.path, 'coords.json'), 'r') as f:
            self.dict_coords = json.load(f)

    def __len__(self):
        return self.len_dataset

    def __getitem__(self, index):
        name_file = self.list_name_file[index]
        path_img = os.path.join(self.path, name_file)

        img = Image.open(path_img)
        coord = self.dict_coords[name_file]

        if self.transform:
            img = self.transform(img)
            coord = torch.tensor(coord, dtype=torch.float32)


        return img, coord
````

#### Создание модели для задачи регресии

````python
class MyModel(nn.Module):
    def __init__(self, input, output):
        super().__init__()
        self.layer_1 = nn.Linear(input, 128)
        self.layer_2 = nn.Linear(128, output)
        self.act = nn.ReLU()

    def forward(self, x):
        x = self.layer_1(x)
        x = self.act(x)
        out = self.layer_2(x)
        return out
````

````python
# создаем и отправляем на cuda
model = MyModel(64*64, 2).to(device)
````
