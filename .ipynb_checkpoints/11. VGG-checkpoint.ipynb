### Теория

**VGG (Visual Geometry Group): Простая и Эффективная Глубокая Сверточная Сеть**

**Основная идея:** VGG - это семейство сверточных нейронных сетей, которые продемонстрировали, что увеличение глубины сети с использованием маленьких фильтров может значительно улучшить производительность в задачах компьютерного зрения.

**Ключевые особенности:**

*   **Однородная архитектура:** VGG отличается простотой и однородностью своей архитектуры. Она состоит из последовательности блоков, каждый из которых включает в себя сверточные слои (с небольшими фильтрами 3x3) и слои макс-пулинга (2x2).
*   **Маленькие фильтры 3x3:** VGG использует только сверточные слои с маленькими фильтрами 3x3. Это позволяет создавать более глубокие сети и захватывать более сложные признаки, чем сети с более крупными фильтрами.
*   **Глубина:** Сети VGG довольно глубокие, с 16-19 слоями (например, VGG16, VGG19), что было значительным шагом вперед по сравнению с более ранними архитектурами.
*   **Сверточные слои с ReLU:** После сверточных слоев обычно используется функция активации ReLU (Rectified Linear Unit).
*   **Полносвязные слои:** После сверточных слоев следуют несколько полносвязных (fully connected) слоев для выполнения классификации.

**Алгоритм работы VGG:**

1.  **Вход:** Изображение подается на вход сети.
2.  **Последовательность сверточных блоков:** Изображение проходит через несколько блоков, состоящих из следующих частей:
    *   **Несколько сверточных слоев (Conv):** Каждый сверточный слой использует фильтры размером 3x3.
    *   **Слой активации ReLU:** Применяется после каждого сверточного слоя.
    *   **Слой макс-пулинга (Max-Pooling):** Уменьшает размер карт признаков (feature maps), сохраняя наиболее важную информацию. Используется с фильтром 2x2.
3.  **Полносвязные слои (Fully Connected):** После сверточных слоев следуют несколько полносвязных слоев. Они используются для преобразования признаков в вектор, который можно использовать для классификации.
4.  **Выходной слой:** Последний полносвязный слой имеет количество нейронов, равное количеству классов. Для классификации используется функция softmax.
5.  **Выход:** На выходе получается вектор с вероятностями классов.

**Варианты VGG:**

*   **VGG16:** 16 слоев (13 сверточных и 3 полносвязных).
*   **VGG19:** 19 слоев (16 сверточных и 3 полносвязных).
*   Существуют также другие варианты, но VGG16 и VGG19 являются наиболее распространенными.
*   Различия в моделях определяются количеством слоев в сверточной части.

**Почему VGG работает?**

*   **Глубина:** Использование нескольких слоев с маленькими фильтрами 3x3 позволяет сети захватывать более сложные и абстрактные признаки, чем более поверхностные сети.
*   **Однородность:** Простая и однородная архитектура упрощает обучение и анализ сети.
*   **Производительность:** VGG продемонстрировала впечатляющие результаты на различных задачах компьютерного зрения.
*   **Универсальность:** VGG можно использовать как основу для решения других задач, таких как обнаружение объектов и сегментация.

**Ограничения VGG:**

*   **Большое количество параметров:** VGG имеет большое количество параметров, что требует значительных вычислительных ресурсов и памяти.
*   **Некоторые модели громоздкие:** Модели VGG могут быть громоздкими, что может сделать их неэффективными на мобильных или встраиваемых устройствах.

**Применение VGG:**

*   Классификация изображений.
*   Обнаружение объектов.
*   Сегментация изображений.
*   Перенос стиля.
*   И множество других задач компьютерного зрения.

**Итог:**

VGG - это архитектура, которая продемонстрировала, что глубокие сети с маленькими сверточными фильтрами могут значительно улучшить производительность в задачах компьютерного зрения. Несмотря на то, что VGG не является самой современной архитектурой, она все еще играет важную роль в области глубокого обучения и является основой для многих других моделей.

### Реализация VGG 11 "в лоб"

Очень легко запутаться.
Обычно в архитектуре VGG используется 5 пулингов

Код ниже определяет класс `MyVGG11`, который представляет собой упрощенную реализацию архитектуры нейронной сети VGG11, известной в области компьютерного зрения. Разберем его по частям:

**1. `__init__(self, num_classes=1000, dropout=0.5)`:**

* **`super().__init__()`:** Вызывает конструктор родительского класса (`nn.Module`), выполняя необходимую инициализацию.
* **`num_classes=1000`:**  Определяет количество выходных классов для классификации. По умолчанию равно 1000, что соответствует набору данных ImageNet.  Можно изменить это значение для задач с другим количеством классов.
* **`dropout=0.5`:**  Задает вероятность применения dropout-регуляризации в полносвязных слоях. Dropout помогает предотвратить переобучение модели.
* **`self.features`:**  Этот атрибут содержит последовательность сверточных слоев (Convolutional Layers), объединенных с функциями активации ReLU (Rectified Linear Unit) и слоями макспулинга (Max Pooling). Эта часть сети отвечает за извлечение признаков из входного изображения.  
    * `nn.Conv2d(...)`: Сверточный слой. Параметры определяют количество входных и выходных каналов, размер ядра свертки и отступы (padding).
    * `nn.ReLU(True)`: Функция активации ReLU, применяемая после каждого сверточного слоя. `True` указывает на использование inplace операции, что экономит память.
    * `nn.MaxPool2d(2)`: Слой макспулинга с размером окна 2x2. Уменьшает размерность карт признаков, сохраняя наиболее важные признаки.  Обратите внимание на структуру VGG11: группы сверточных слоев с одинаковым количеством фильтров, за которыми следует слой макспулинга.
* **`self.avgpool = nn.AdaptiveAvgPool2d((7, 7))`:** Слой адаптивного усредняющего пулинга. Он преобразует выход `self.features` в тензор с фиксированным размером 7x7, независимо от входного размера. Это позволяет использовать сеть с изображениями разного разрешения.
* **`self.flatten = nn.Flatten()`:** Преобразует многомерный тензор из `self.avgpool` в одномерный вектор,  подготавливая данные для полносвязных слоев.
* **`self.classifier`:**  Этот атрибут содержит последовательность полносвязных слоев (Fully Connected Layers), также с ReLU и Dropout. Эта часть сети отвечает за классификацию на основе извлеченных признаков.
    * `nn.Linear(...)`: Полносвязный слой. Параметры определяют количество входных и выходных нейронов.
    * `nn.Dropout(p=dropout)`: Слой Dropout с заданной вероятностью отключения нейронов.


**2. `forward(self, x)`:**

* Эта функция описывает прямой проход данных через сеть. Она принимает входной тензор `x` (изображение) и возвращает тензор `out` с предсказаниями вероятностей для каждого класса.
* Последовательно применяет все слои, определенные в `__init__`:  `self.features`, `self.avgpool`, `self.flatten` и `self.classifier`.

```python
class MyVGG11(nn.Module):
    def __init__(self, num_classes=1000, dropout=0.5):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, (3, 3), padding=1),
            nn.ReLU(True),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, (3, 3), padding=1),
            nn.ReLU(True),
            nn.MaxPool2d(2),
            nn.Conv2d(128, 256, (3, 3), padding=1),
            nn.ReLU(True),
            nn.Conv2d(256, 256, (3, 3), padding=1),
            nn.ReLU(True),
            nn.MaxPool2d(2),
            nn.Conv2d(256, 512, (3, 3), padding=1),
            nn.ReLU(True),
            nn.Conv2d(512, 512, (3, 3), padding=1),
            nn.ReLU(True),
            nn.MaxPool2d(2),
            nn.Conv2d(512, 512, (3, 3), padding=1),
            nn.ReLU(True),
            nn.Conv2d(512, 512, (3, 3), padding=1),
            nn.ReLU(True),
            nn.MaxPool2d(2)
        )
        
        self.avgpool = nn.AdaptiveAvgPool2d((7, 7))

        self.flatten = nn.Flatten()

        self.classifier = nn.Sequential(
            nn.Linear(512 * 7 * 7, 4096),
            nn.ReLU(True),
            nn.Dropout(p=dropout),
            nn.Linear(4096, 4096),
            nn.ReLU(True),
            nn.Dropout(p=dropout),
            nn.Linear(4096, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.avgpool(x)
        x = self.flatten(x)
        out = self.classifier(x)
        return out
```

```python
# инициализируем модель
model = MyVGG11()
model
```

```python
# проверим сборку модели на корректность
input = torch.rand([1, 3, 224, 224], dtype=torch.float32)

pred = model(input)
print(pred.shape)
```

### Реализация общего класса VGG

Код ниже определяет класс `MyVGG11`, который представляет собой упрощенную реализацию архитектуры нейронной сети VGG11, известной в области компьютерного зрения. Разберем его по частям:

    * `nn.Linear(...)`: Полносвязный слой. Параметры определяют количество входных и выходных нейронов.

**2. `forward(self, x)`:**

* Последовательно применяет все слои, определенные в `__init__`:  `self.features`, `self.avgpool`, `self.flatten` и `self.classifier`.


```python
cfgs = {
    "vgg_11":[64,"M", 128, "M", 256, 256, "M", 512, 512, "M", 512, 512, "M"],
    "vgg_13":[64, 64,"M", 128, 128, "M", 256, 256, "M", 512, 512, "M", 512, 512, "M"],
    "vgg_16":[64, 64,"M", 128, 128, "M", 256, 256, 256, "M", 512, 512, 512, "M", 512, 512, 512, "M"],
    "vgg_19":[64, 64,"M", 128, 128, "M", 256, 256, 256, 256, "M", 512, 512, 512, 512, "M", 512, 512, 512, 512, "M"]
}
```

```python
class MyVGG(nn.Module):
    _cfgs = {
        "vgg_11":[64,"M", 128, "M", 256, 256, "M", 512, 512, "M", 512, 512, "M"],
        "vgg_13":[64, 64,"M", 128, 128, "M", 256, 256, "M", 512, 512, "M", 512, 512, "M"],
        "vgg_16":[64, 64,"M", 128, 128, "M", 256, 256, 256, "M", 512, 512, 512, "M", 512, 512, 512, "M"],
        "vgg_19":[64, 64,"M", 128, 128, "M", 256, 256, 256, 256, "M", 512, 512, 512, 512, "M", 512, 512, 512, 512, "M"]
    }

    def __init__(self, model_name, num_classes=1000, dropout=0.5):
        super().__init__()
        self.cfg = self._cfgs[model_name]
        self.features = self.make_layers(self.cfg)
        self.avgpool = nn.AdaptiveAvgPool2d((7, 7))
        self.flatten = nn.Flatten()
        
        self.classifier = nn.Sequential(
            nn.Linear(512 * 7 * 7, 4096),
            nn.ReLU(True),
            nn.Dropout(p=dropout),
            nn.Linear(4096, 4096),
            nn.ReLU(True),
            nn.Dropout(p=dropout),
            nn.Linear(4096, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.avgpool(x)
        x = self.flatten(x)
        out = self.classifier(x)
        return out

    def make_layers(self, cfg):
        layers = []
        in_channels = 3
        for value in cfg:
            if value == "M":
                layers += [nn.MaxPool2d(2)]
            else:
                conv2d = nn.Conv2d(in_channels, value, (3,3), padding=1)
                layers += [conv2d, nn.ReLU(True)]
                in_channels = value
        return nn.Sequential(*layers)
```

### Предсказания обученной модели

```python
img = Image.open('test_pic_1.png')
```

```python
plt.imshow(img)
```

```python
# Дефолтные преобразования
transform_default = weights_vgg16.transforms()
transform_default
```

```python
# Преобразования изображения на основе модуля V1
size = 224

transform_v1 = v1.Compose([
    v1.Resize(256),
    v1.CenterCrop(size),
    v1.ToTensor(),
    v1.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225))
])
```

```python
# Преобразования изображения на основе модуля V2
size = 224

transform_v2 = v2.Compose([
    v2.ToImage(),
    v2.Resize(256),
    v2.CenterCrop(size),
    v2.ToDtype(dtype=torch.float32, scale=True),
    v2.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225))
])
```

```python
transforms = {
    'transform_default': transform_default,
    'transform_v1':transform_v1,
    'transform_v2':transform_v2
}

plt.imshow(img)
plt.show()

vgg16.eval()

for name, transform in transforms.items():
    print(f'{name}:')
    img_in = transform(img).unsqueeze(dim=0)    # img => transform => (3, 224, 224) => unsqueeze => (1, 3, 224, 224) => img_in
    pred = vgg16(img_in).squeeze()    # (1, 1000) => (1000)

    sorted, indices = pred.softmax(dim=0).sort(descending=True)
    for i, (s, ind) in enumerate(zip(sorted, indices)):
        print(f'    {class_names[ind]} => {s:.5f}')
        if i // 5:
            break
    print('===================', end='\n\n')
```
